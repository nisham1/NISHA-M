# LIST
<p style="font-family: Serif ; font-size:1.4em;color:BLACK;"> List is mutable and ordered in nature , we can access the elements of list. using their index.Index starts from 0 and ranges till length-1
Index can also be negative values for instance the last element in the list can be accessed using index -1 and the first number is accessed using index -length .We can perform various operations on list like slicing,sorting,insert and so on.
<p/>

<img src="https://www.aipython.in/wp-content/uploads/2020/02/Index_in_python-list.png" />

```python
list_variable=[1,2,3,4]
print(list_variable[0],"Before manipulation")
list_variable[0]='Zero'
print(list_variable,"After manipulation")
```
