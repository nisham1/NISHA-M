```python
#ACTIVATION FUNCTION 1
#BINARY STEP FUNCTION OR THRESHOLD FUNCTION
import matplotlib.pyplot as plt
def f(x):
    if(x<0):
        return 0
    return 1
inputs=[]
outputs=[]
for i in range(-1,5):
    inputs.append(i)
    outputs.append(f(i))
plt.plot(inputs,outputs)
plt.title("THRESHOLD FUNCTION / BINARY STEP FUNCTION")
plt.xlabel("input".upper())
plt.ylabel("output".upper())


#THIS FUNCTION PROVIDES ONLY TWO VALUES EITHER ZERO OR ONE AND CAN BE USED FOR BINARY CLASSIFICATION PROBLEM
#IF THE PROVIDED INPUT VALUE IS LESS THAN THE THRESHOLD VALUE THE ACTIVATION FUNCTION RETURNS ZERO
#IF THE PROVIDED INPUT VALUE IS GREATER THAN THE THRESHOLD VALUE THE ACTIVATION FUNCTION RETURNS ONE
#HERE WE HAVE ASSUMED THRESHOLD VALUE AS 0
```

```python
#LOSS FUNCTION-1
#MEAN SQUARED ERROR
def mse(actual,predicted):
    s=0
    for i in range(len(actual)):
        s+=(actual[i]-predicted[i])**2
    return(1/(len(actual)))*s
inputs=[1,2,3,4]
outputs=[2,3,4,5]

mse(inputs,outputs)
```

```python
#LOSS FUNCTION-2
#MEAN ABSOLUTE ERROR / L1 LOSS FUNCTION
import matplotlib.pyplot as plt
import numpy as np
ypred = np.array([0.000, 0.166, 0.333])
yact = np.array([0.000, 0.254, 0.998])

print("d is: " + str(["%.8f" % elem for elem in ypred]))
print("p is: " + str(["%.8f" % elem for elem in yact]))

def mae(predictions, targets):
    differences = predictions - targets
    absolute_differences = np.absolute(differences)
    mean_absolute_differences = absolute_differences.mean()
    return mean_absolute_differences
mae_val = mae(ypred, yact)

print ("mae error is: " + str(mae_val))
```

```python
#loss function-3
#MEAN BIAS ERROR
#mbe=(summation(i=1 to n)(y-yhat))/n
def mbe(actual,predicted):
    s=0
    for i in range(len(actual)):
        s+=(actual[i]-predicted[i])
    return(1/(len(actual)))*s
inputs=[1,2,3,4]
outputs=[2,3,4,5]
mbe(inputs,outputs)
```

```python
#CROSS ENTROPY LOSS OR LOGRATHMIC LOSS OR  BINARY CROSS ENTROPY
#LOSS FUNCTION-4
from math import log
def binary_cross_entropy(actual,predicted):
    s=0
    for i in range(len(actual)):
        s+=(actual[i]*log(1e-15+predicted[i]))
    return -(1/(len(actual)))*s
inputs=[1,2,3,4]
outputs=[2,3,4,5]
binary_cross_entropy(inputs,outputs)
```

```python
#CATEGORICAL CROSS ENTROPY
#LOSS FUNCTION-5

from math import log
def cross_entropy(predictions, targets, epsilon=1e-10):
    predictions = np.clip(predictions, epsilon, 1. - epsilon)
    N = predictions.shape[0]
    ce_loss = -np.sum(np.sum(targets * np.log(predictions + 1e-5)))/N
    return ce_loss
inp=[[1,2,3],[4,5,6],[7,8,9]]
out=[[2,3,4],[5,6,7],[8,9,10]]
cross_entropy(inp,out)
```

```python
#LOSS FUNCTION-6
#HINGE LOSS  =max(0,1-y*hinge(x))  
def update_weights_Hinge(m1, m2, b, X1, X2, Y, learning_rate):
    m1_deriv = 0
    m2_deriv = 0
    b_deriv = 0
    N = len(X1)
    for i in range(N):
        # Calculate partial derivatives
        if Y[i]*(m1*X1[i] + m2*X2[i] + b) <= 1:
          m1_deriv += -X1[i] * Y[i]
          m2_deriv += -X2[i] * Y[i]
          b_deriv += -Y[i]
        # else derivatives are zero

    # We subtract because the derivatives point in direction of steepest ascent
    m1 -= (m1_deriv / float(N)) * learning_rate
    m2 -= (m2_deriv / float(N)) * learning_rate
    b -= (b_deriv / float(N)) * learning_rate

    return m1, m2, b

```

```python
#LOSS FUNCTION-7
#HUBER LOSS FUNCTION
def update_weights_Huber(m, b, X, Y, delta, learning_rate):
    m_deriv = 0
    b_deriv = 0
    N = len(X)
    for i in range(N):
        # derivative of quadratic for small values and of linear for large values
        if abs(Y[i] - m*X[i] - b) <= delta:
          m_deriv += -X[i] * (Y[i] - (m*X[i] + b))
          b_deriv += - (Y[i] - (m*X[i] + b))
        else:
          m_deriv += delta * X[i] * ((m*X[i] + b) - Y[i]) / abs((m*X[i] + b) - Y[i])
          b_deriv += delta * ((m*X[i] + b) - Y[i]) / abs((m*X[i] + b) - Y[i])
    
    # We subtract because the derivatives point in direction of steepest ascent
    m -= (m_deriv / float(N)) * learning_rate
    b -= (b_deriv / float(N)) * learning_rate

    return m, b
```
